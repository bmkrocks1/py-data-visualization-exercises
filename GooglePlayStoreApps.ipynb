```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt

import plotly
import plotly.graph_objs as go

# This injects the plotly.js source file into the notebook
plotly.offline.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv("datasets/googleplaystore.csv")

# Take a peek at the first five rows in the dataset
df.head()
```

```python
# Clean up the dataframe from duplicate Apps
df.drop_duplicates(subset='App', inplace=True)

# Filter out rows with Installs equal to 'Free'
df = df[df['Installs'] != 'Free']

# Installs -- Remove '+' and ','
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)

# Size -- Replace 'Varies with device' with NaN
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)

# Size -- Convert all app sizes to MB
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(x))

# Price -- Remove '$'
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
df['Price'] = df['Price'].apply(lambda x: float(x))

df['Reviews'] = df['Reviews'].apply(lambda x: int(x))
```

```python
# Plot data using plotly Histogram
data = [go.Histogram(
    x = df['Rating'],
    xbins = dict(start=1, size=.1, end=5)
)]

plotly.offline.iplot(data, filename='overall_rating_distribution')
```
